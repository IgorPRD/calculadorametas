# **META ADS**



---



Utilize o CPM médio das campanhas ativas do cliente ou de um cliente de setor similar. Se não houver a possibilidade de extrair o CPM, insira o público planejado dentro de uma campanha real no Meta ADS, pegue o menor valor de alcance diário disponibilizado na plataforma e insira a métrica na segunda calculadora da seção de META ADS, que executará o cálculo de acordo com o alcance.

---



```python
#@title Calculadora de Metas - Meta ADS - [CPM]

import math
import numpy as np
import pandas as pd

setor = "agencia de marketing e publicidade" #@param ["software e cloud", "ecommerce", "saude e estetica", "engenharia e industria geral", "servicos em geral", "agencia de marketing e publicidade"]
cpm = 57.28 #@param {type:"number"}
segmento = "b2b" #@param ["b2b", "b2c"]
verba = 600 #@param {type:"number"}
objetivo = "material rico" #@param ["fundo de funil", "material rico"]

software = "software e cloud"
ecommerce = "ecommerce"
saude = "saude e estetica"
industria = "engenharia e industria geral"
servicos = "servicos em geral"
agencia = "agencia de marketing e publicidade"
bofu = "fundo de funil"
mofu = "material rico"

if setor == software:
  vl = 5.70
  lo = 12.20
  ov = 14.8
elif setor == ecommerce:
  vl = 3
  lo = 21.10
  ov = 61.40
elif setor == saude:
  vl = 5.10
  lo = 22.40
  ov = 36.30
elif setor == industria:
  vl = 3.8
  lo = 17.60
  ov = 37.30
elif setor == servicos:
  vl = 4.10
  lo = 22.70
  ov = 35.80
elif setor == agencia:
  vl = 4
  lo = 13.50
  ov = 24


if segmento == 'b2b':
  ctr = 0.7
if segmento == 'b2c':
  ctr = 0.35
if setor == ecommerce:
  ctr = 1

if objetivo == bofu:
  lo = 100


impressoes = ((verba/cpm) * 1000)
cliques = ((ctr*impressoes) / 100)
lead = ((cliques*vl) / 100)
oportunidade = ((lead*lo) / 100)
venda = ((oportunidade*ov) / 100)

index = {0: 'Impressões', 1: 'Cliques', 2: 'Leads', 3: 'Oportunidades', 
         4: 'Vendas'}

resultados = np.array([round(impressoes), round(cliques), math.floor(lead),
          math.floor(oportunidade), round(venda)])

resultados_df = pd.DataFrame(resultados)

resultados_df.columns = ['Estimativa']
resultados_final = resultados_df.rename(index=index)

resultados_final
```

# **GOOGLE ADS**



---



Insira todas as palavras-chave selecionadas na criação de campanha de pesquisa no Google ADS e use a previsão de CPC médio diponibilizada para executar o cálculo.

```python
#@title Calculadora de Metas - Google ADS

import math
import numpy as np
import pandas as pd

setor = "software e cloud" #@param ["software e cloud", "ecommerce", "saude e estetica", "engenharia e industria geral", "servicos em geral", "agencia de marketing e publicidade"]
cpc = 0.6 #@param {type:"number"}
verba = 1200 #@param {type:"number"}
objetivo = "material rico" #@param ["fundo de funil", "material rico"]

software = "software e cloud"
ecommerce = "ecommerce"
saude = "saude e estetica"
industria = "engenharia e industria geral"
servicos = "servicos em geral"
agencia = "agencia de marketing e publicidade"
bofu = "fundo de funil"
mofu = "material rico"

if setor == software:
  vl = 5.70
  lo = 12.20
  ov = 14.8
elif setor == ecommerce:
  vl = 3
  lo = 21.10
  ov = 61.40
elif setor == saude:
  vl = 5.10
  lo = 22.40
  ov = 36.30
elif setor == industria:
  vl = 3.8
  lo = 17.60
  ov = 37.30
elif setor == servicos:
  vl = 4.10
  lo = 22.70
  ov = 35.80
elif setor == agencia:
  vl = 4
  lo = 13.50
  ov = 24

if objetivo == bofu:
  lo = 100

cliques = (verba / cpc)
lead = ((cliques*vl) / 100)
oportunidade = ((lead*lo) / 100)
venda = ((oportunidade*ov) / 100)

resultados = np.array([round(cliques), math.floor(lead), 
                       math.floor(oportunidade), round(venda)])

resultados_df = pd.DataFrame(resultados)

index = {0: 'Cliques', 1: 'Leads', 2: 'Oportunidades', 3: 'Vendas'}

resultados_df.columns = ['Estimativa']

resultados_final = resultados_df.rename(index=index)

resultados_final
```

# **META DO CICLO / ROAS**



---



Some as métricas estimadas de todas as campanhas planejadas para executar o cálculo. Visto que cada ciclo tem 3 meses, lembre-se de multiplicar os resultados por 3 antes de inserí-los na calculadora.



---



```python
#@title Calculadora de Meta do Ciclo / ROAS

import math
import numpy as np
import pandas as pd

usuarios = 17478 #@param {type:"number"}
leads = 975 #@param {type:"number"}
oportunidades = 642 #@param {type:"number"}
vendas = 44 #@param {type:"number"}
ticket_medio = 600 #@param {type:"number"}
verba = 20590 #@param {type: "number"}

cpc = (verba / usuarios)
cpl = (verba / leads)
cpo = (verba / oportunidades)
cac = (verba / vendas)
faturamento = (ticket_medio * vendas)
roas = (((faturamento - verba) / verba) * 100)


resultados = np.array([round(faturamento), round(cpc, 2), round(cpl, 2),
                       round(cpo, 2), round(cac, 2), round(roas, 2)])

resultados_df = pd.DataFrame(resultados)

index = {0: 'Faturamento', 1: 'CPC', 2: 'CPL', 3: 'CPO', 4: 'CAC', 
         5: 'ROAS'}

resultados_df.columns = ['Estimativa']

resultados_final = resultados_df.rename(index=index)

resultados_final
```

Se a estimativa de resultados gerada com a verba dispobilizada não atingir a meta de resultados pré-estabelecida pelo cliente, utilize a calculadora abaixo para verificar a verba necessária para atingir a meta.

```python
#@title Calculadora de Previsão de Verba Necessária Para Meta Pré-definida

import math
import numpy as np
import pandas as pd

resultado_atual = 10 #@param {type:"number"}
verba_atual = 5000 #@param {type:"number"}
meta = 20 #@param {type:"number"}

verba = ((meta*verba_atual) / resultado_atual)

resultado = np.array([verba])

resultado_df = pd.DataFrame(resultado)

index = {0: 'Verba Necessária'}

resultado_df.columns = ['Estimativa']
resultado_final = resultado_df.rename(index=index)


resultado_final
```
